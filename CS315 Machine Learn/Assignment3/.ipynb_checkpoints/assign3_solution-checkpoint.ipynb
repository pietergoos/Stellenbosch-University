# Do it yourself

## Generate Data

```python
# Training data - two randomly-generated Gaussian-distributed clouds of points in 2d space
np.random.seed(0)
# Number of points
N = 1000
# Labels for each cluster
y = np.random.randint(low=0, high=2, size = N)
c = np.array(["r","g"])
# Mean of each cluster
means = np.array([[-2, 2], [-2, 2],])
# Covariance (in X and Y direction) of each cluster
covariances = np.random.random_sample((2, 2)) + 1
# Dimensions of each point
X = np.vstack([np.random.randn(N)*covariances[0, y] + means[0, y],
               np.random.randn(N)*covariances[1, y] + means[1, y]])

for k in range(X.shape[1]):
    plt.plot(X[0,k],X[1,k],c[y[k]]+"o")
    
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.title("Class labels given")
plt.show()

for k in range(X.shape[1]):
    plt.plot(X[0,k],X[1,k],"m"+"o")
    
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.title("No Class labels")
plt.show()
```

## K-means

```python
#RUN K-MEANS
B=gkroon.kmeans(data=X, K=2)
mean = B[0]
label= B[1]
for k in range(X.shape[1]):
    plt.plot(X[0,k],X[1,k],c[label[k,0]]+"o")
plt.plot(mean[0,0],mean[1,0],"ks")
plt.plot(mean[0,1],mean[1,1],"cs")
plt.xlabel("$x_1")
plt.ylabel("$x_2")
plt.show()
```

## GMM

```python
#RUN GMM
B = gkroon.gmm(data=X,K=2,hard=False)
#print(B[0]) components
mean = B[1]
cov = B[2]
#print(mean[:,0])#klas 0 se mean
#print(mean[:,1])#klas 1 se mean
#print(cov[0,:,:])#klas 0 se cov
#print(cov[1,:,:])#klas 1 se cov

R = gkroon.calcresps(data=X, nums=B[0], means=B[1], covs=B[2], hard=False)
resp = R[0]
#print(R[0].shape)
#print(X.shape)
```

```python
#Function to help with plotting confidence interval
def eigsorted(cov):
    vals, vecs = np.linalg.eigh(cov)
    order = vals.argsort()[::-1]
    return vals[order], vecs[:,order]
```

```python
ax = plt.gca()
nstd = 2.0
##ELLIPSE 1
vals, vecs = eigsorted(cov[0,:,:])
theta = np.degrees(np.arctan2(*vecs[:,0][::-1]))
w, h = 2 * nstd * np.sqrt(vals)
ell1 = Ellipse(xy=(mean[0,0], mean[1,0]),
              width=w, height=h,
              angle=theta, edgecolor='green',facecolor='white',fill=True,linewidth=3,zorder=1)
ell1.set_facecolor('none')
ax.add_artist(ell1)

##ELLIPSE 2
vals, vecs = eigsorted(cov[1,:,:])
theta = np.degrees(np.arctan2(*vecs[:,0][::-1]))
w, h = 2 * nstd * np.sqrt(vals)
ell2 = Ellipse(xy=(mean[0,1], mean[1,1]),
              width=w, height=h,
              angle=theta, edgecolor='red',facecolor='white',fill=True,linewidth=3,zorder=1)
ell2.set_facecolor('none')
ax.add_artist(ell2)


for i in range(X.shape[1]):
         col = [[resp[i,1],resp[i,0],0]]
         ax.scatter(X[0,i],X[1,i],c=col,zorder=3,alpha=0.2)
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")
plt.xlim([-6,6])
plt.ylim([-6,6])
plt.show()
```

# Perform $k$-means on an image

```python
im = io.imread('./cartoon.png')
plt.imshow(im/np.max(im))
plt.show()
```

```python
m,n = im.shape[:2]
data = im.reshape(m*n,3)
data = np.array(data, dtype = float)

n_clusters = 3
km = KMeans(n_clusters)
km.fit(data)
labels = km.labels_
clus_cntrs = km.cluster_centers_



dat = np.zeros(data.shape)

for i in range(n_clusters):
    ind = (labels == i)
    mn = clus_cntrs[i,:]
    
    mn = mn[np.newaxis,:]
    dat[ind,:] = mn


dat = dat.reshape(m,n,3)
dat = np.asarray(dat,dtype="uint8")


plt.imshow(dat)
plt.show()
```

## GMM

Finally we fit a GMM to the digit data. 

```python
data = digits.data

np.random.seed(1)
g = GaussianMixture(n_components=10,covariance_type = 'full')

g.fit(data)
prob = g.predict_proba(data)
n,k = prob.shape



mns = g.means_
covs = g.covariances_
im = mns.reshape(10,8,8)
def plot_means(k=0):
    plt.imshow(im[k], cmap = cm.binary)
    plt.show()
```

```python
interact(plot_means,k=(0,9))
```

```python
# Figure size in inches
fig = plt.figure(figsize=(8, 3))

# Add title
fig.suptitle('Cluster Center Images', fontsize=14, fontweight='bold')

# For all labels (0-9)
for i in range(10):
    # Initialize subplots in a grid of 2X5, at i+1th position
    ax = fig.add_subplot(2, 5, 1 + i)
    # Display images
    ax.imshow(im[i].reshape((8, 8)), cmap=plt.cm.binary)
    # Don't show the axes
    plt.axis('off')

# Show the plot
plt.show()
```
